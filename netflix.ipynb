```python
#Посмотреть уникальные значения в столбце 'type':
df['type'].unique()
#Посмотреть количество уникальных значений в столбце 'type':
df['type'].value_counts()
```

```python
#Делаем фильтр только для фильмов:
movies = df[df['type'] == 'Movie']
#Делаем фильтр только для сериалов:
serials = df[df['type'] == 'TV Show']
#Смотрю кол-во фильмов и сериалов:
print(f'Всего фильмов: {len(movies)}')
print(f'Всего сериалов: {len(serials)}')
```

```python
#Фильмы по годам:
minimal_year = df['release_year'].min()
maximal_year = df['release_year'].max()
media_year = df['release_year'].median()
print(f'Самый старый фильм в {minimal_year} году')
print(f'Самый свежий фильм в {maximal_year} году')
print(f'Медиана года выпуска фильмов: {media_year}')
```
